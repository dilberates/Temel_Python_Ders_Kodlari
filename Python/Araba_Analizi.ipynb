```python
y=dframe["Type"].values
x=dframe.drop("Type",axis=1).values
```

```python
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
scaler1=MinMaxScaler()
```

```python
scaler1.fit(x_train)
```

```python
x_train=scaler1.transform(x_train)
x_test=scaler1.transform(x_test)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model1=Sequential()

model1.add(Dense(units=30,activation="relu")) #kaç özellik varsa giriş nöron sayısını o yapmamız tavsiye edilir 
model1.add(Dense(units=15,activation="relu"))
model1.add(Dense(units=15,activation="relu"))
model1.add(Dense(units=1,activation="relu"))

model1.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model1.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1)
```

```python
modelKayip=pd.DataFrame(model1.history.history)
```

```python
modelKayip.plot()
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700, validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(30,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(15,activation="sigmoid"))

model.add(Dense(1))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700, validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
kayipdf=pd.DataFrame(model.history.history)
```

```python
kayipdf.plot()
```

```python
tahminler=model.predict_classes(x_test)
```

```python
tahminler
```

```python
 from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahminler))
```

```python
print(confusion_matrix(y_test,tahminler))
```
