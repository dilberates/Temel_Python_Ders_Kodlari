```python
numpyDizi1=np.linspace(0,10,20)
```

```python
numpyDizi1
```

```python
numpyDizi2=numpyDizi1 ** 3
```

```python
numpyDizi2
```

```python
plt.plot(numpyDizi1,numpyDizi2,"pink")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizi1,numpyDizi2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyDizi2,numpyDizi1,"g--")
```

```python
benimFigure=plt.figure()
figureAxes=benimFigure.add_axes([0.4,0.4,0.6,0.6])
figureAxes.plot(numpyDizi1,numpyDizi2,"black")
figureAxes.set_xlabel("X ekseni başlık")
figureAxes.set_ylabel("Y ekseni başlık")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.6,0.9,0.9])

eksen1.plot(numpyDizi1,numpyDizi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Başlık")

eksen2.plot(numpyDizi2,numpyDizi1,"pink")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Ana Başlık")
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)

for eksen in benimEksenler:
    eksen.plot(numpyDizi1,numpyDizi2,"b")
    eksen.set_xlabel("x EKSENİ")

plt.tight_layout()
```

```python
yeniFigure=plt.figure(dpi=200)

yeniEksen=yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizi1,numpyDizi1 - 2,label="numpyDizi - 2")

yeniEksen.plot(numpyDizi1,numpyDizi1 * 3,label="numpyDizi * 3")

yeniEksen.legend(loc=0)
```

```python
yeniFigure.savefig("benimfigur.png",dpi=200)
```
