```python
import numpy as np
import matplotlib.pyplot as pyt
%matplotlib inline
```

```python
numpyDizi=np.linspace(0,10,20)
```

```python
numpyDizi
```

```python
numpyDizi2=numpyDizi**2
```

```python
(benimFigure,benimEksen)=pyt.subplots()
benimEksen.plot(numpyDizi,numpyDizi2,color="#B3F02F",alpha=1)
benimEksen.plot(numpyDizi2,numpyDizi,color="#48432F")
```

```python
(yeniFigur,yeniEksen)=pyt.subplots()
yeniEksen.plot(numpyDizi,numpyDizi+2,color="blue",linewidth=3)
yeniEksen.plot(numpyDizi,numpyDizi+3,color="pink",linestyle=":")
yeniEksen.plot(numpyDizi,numpyDizi+4,color="red",linestyle="--")
yeniEksen.plot(numpyDizi,numpyDizi+5,color="yellow",linestyle="-.")
yeniEksen.plot(numpyDizi,numpyDizi+6,color="black",linestyle="--",marker="o",markersize=6,markerfacecolor="red")
yeniEksen.plot(numpyDizi,numpyDizi+7,color="black",linestyle="--",marker="*",markersize=8,markerfacecolor="red")
```
